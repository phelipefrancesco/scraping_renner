```python
def seila(x):
    
    try:
        
        return geolocator.geocode(x)
    
    except:
        return None
    
def seila2(x):
    
    try:
        
        return geolocator.geocode(x).latitude
    
    except:
        return None

def seila3(x):
    
    try:
        
        return geolocator.geocode(x).longitude
    
    except:
        return None
```
